```python
from bs4 import BeautifulSoup
import urllib.request, urllib.parse, urllib.error


import pandas as pd
```

```python
links_to_places = list()
base_url = "https://en.wikipedia.org"
print(urllib.request.urljoin(base_url,"wiki/Lists_of_tourist_attractions"))
html = urllib.request.urlopen(urllib.request.urljoin(base_url,"wiki/Lists_of_tourist_attractions")).read()

parsed = BeautifulSoup(html, 'html.parser')
```

```python
country_header = parsed.find('span', text = 'By country', class_ = 'mw-headline').parent

for country in country_header.find_next_siblings('ul'):
    try:
        country_name = country.find_previous_sibling('h3').get_text().split('[')[0]
        for links in country.find_all('li'):
            link = links.find('a')['href']
            title = links.find('a')['title']
            links_to_places.append((link, title, country_name))
    except:
        pass
```

```python
links_to_places
```

```python
places = set()
remove = set()
res = pd.DataFrame()
for link, title, country_name in links_to_places:
    try:
        html = urllib.request.urlopen(urllib.request.urljoin(base_url, link)).read()
        parsed = BeautifulSoup(html, 'html.parser')
        for tourist_places in parsed.find_all('h3'):
            place = tourist_places.find('span', class_ = 'mw-headline').get_text()
            country = country_name
            if(place.lower() in places):
                remove.add(place.lower())
            res = res.append(pd.DataFrame({'Place' : [place], 'Country' : [country_name], "Info" : [urllib.request.urljoin(base_url, link)]}))
            places.add(place.lower())
    except Exception as e:
        print(e)
        
res = res[~res['Place'].isin(remove)]
```

```python
res.to_csv('list_of_tourist_attractions.tsv', index=False, encoding='utf-16', sep='\t')
```
